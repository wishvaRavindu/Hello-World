```python
import numpy as np
from matplotlib import pyplot as plt


#creating a random image
random_image = np.random.random([500,500])

plt.imshow(random_image, cmap = "gray")
plt.colorbar()
```

```python
from skimage import data


coins  = data.coins()

print("type : ",type(coins))
print("dype : ",coins.dtype)
print("shape : ",coins.shape)

plt.imshow(coins,cmap="gray")
```

```python
cat = data.chelsea()

print("chape",cat.shape)

plt.imshow(cat)
```

```python
#grabbing a set rows and columns from the image and chaning the values of the rgb values

cat[10:110,10:110, :] = [255,0,0]  #[r,g,b]

plt.imshow(cat)
```

```python
img0 = data.chelsea()
img1 = data.rocket()

f,(ax0,ax1) = plt.subplots(1,2,figsize = (20,10))

ax0.imshow(img0)
ax0.set_title("cat",fontSize = 18)
ax0.axis('off')

ax1.imshow(img1)
ax1.set_title('Rocket',fontSize = 18)
ax1.set_xlabel(r'launching position $\alpha = 320$')

ax1.vlines([202,300],0,img1.shape[0],colors = 'magenta',linewidth = 3,label = 'Side to tower postion')
ax1.plot([168,190,200],[400,200,300],color = 'white',linestyle = '--',label = 'Side tower postion')
ax1.legend();
```

```python
from skimage import img_as_float, img_as_ubyte

image  = data.chelsea()

image_ubyte = img_as_ubyte(image)
image_float = img_as_float(image)

print("type, min, max :",image_ubyte.dtype,image_ubyte.min,image_ubyte.max)
print("type, min, max :",image_float.dtype,image_float.min,image_float.max)
```
